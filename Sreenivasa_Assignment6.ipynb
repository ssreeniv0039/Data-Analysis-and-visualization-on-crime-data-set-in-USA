```python
crime_rate = data.groupby(['Population1'])['Violent \ncrime \nrate '].sum()
```

```python
# Plotting graph
data.plot(x = "Population1", y = "Violent crime rate", label = "Violent crime rate")
plt.xlabel("Population")
plt.ylabel("Violent Crime Rate")
plt.legend()
plt.title("Violent crime rate in relation to population size")
plt.show()
```
